```python
def readCoherence(coherence_dict):
    output = pd.DataFrame()
    
    for keys in coherence_dict:
        output.loc[keys, 'Banyak Topik'] = len(coherence_dict[keys])
        output.loc[keys, 'Skor Koherensi'] = np.mean(coherence_dict[keys])
        output.loc[keys, 'Median'] = np.median(coherence_dict[keys])
    
    output = output.round(2)
    output['Banyak Topik'] = output['Banyak Topik'].astype(int)
    
    return output
```

```python
def readTopic(save_name):
    topic_columns_list = []
    
    for fold_num in range(5):
        topic = pd.read_csv(f"{save_name}_fold_{fold_num}_test.csv")
        topic_columns_list.append(topic.columns)
    
    return topic_columns_list
```

## Classification

```python
def readScore(save_name, the_mean=True):
    print(save_name)
    
    score = pd.read_csv(f"{save_name}_score.csv", index_col=0)
    score['data'] = pd.Series(score.index).apply(lambda idx: str(idx)[:-14]).to_list()
    score['fold'] = pd.Series(score.index).apply(lambda idx: str(idx)[-13:-7]).to_list()
    score['resample'] = [1 if 'resampled' in data else 0 for data in score['data']]
    
    if the_mean:
        return score.groupby('data').mean().round(2).loc[['A', 'B', 'C', 'A_resampled', 'B_resampled', 'C_resampled'], ['LogisticRegression', 'LinearDiscriminantAnalysis', 'KNeighborsClassifier', 'RandomForestClassifier']]
    else:
        return score
```
